```python
import pandas as pd

# Spécifiez le chemin complet vers le fichier CSV
chemin_fichier_csv = "/content/drive/My Drive/Télémedecine/lou_total.csv"

# Lisez le fichier CSV en spécifiant le délimiteur comme '\t' (tabulation)
data = pd.read_csv(chemin_fichier_csv, delimiter="\t", header=None)

# Sélectionnez uniquement les colonnes 2, 3 et 4
data = data.iloc[:, 2:5]

# Renommez les colonnes
data.columns = ['THORAX', 'X', 'Y']

# Affichez le DataFrame mis à jour
print(data)
```

```python
# Calcul des fonctions de transfert
# Appliquer la fonction de transfert à chaque valeur de la colonne "X"
data['X'] = data['X'].apply(lambda x: (((x / (2**16 - 1)) - 0.5) * 3))

# Appliquer la fonction de transfert à chaque valeur de la colonne "Y"
data['Y'] = data['Y'].apply(lambda y: (((y / (2**16 - 1)) - 0.5) * 3))

print(data)
```
